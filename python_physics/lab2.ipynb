```python
#s=sumx^n
x=0.25

def summation(x,N):
    sum = 0 
    for i in range(N+1):
        sum = sum + x**i
        #print(sum,'--',i)
    return sum
```

```python
print('for N= 10 sum :',summation(x,10))
print('for N= 100 sum :',summation(x,100))
print('for N= 1000 sum :',summation(x,1000))
```
