## Exercise 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)

h = np.array(10.0**np.arange(-1,-11,-1))
h
```

```python
x = np.pi/4
dfdxf = (f(x+h)-f(x))/h
print(dfdxf)
```

```python
difference = dfdxf - np.cos(x)
print(difference)
```

```python
%matplotlib inline
plt.loglog(h,abs(difference),'o')
plt.xlabel('h')
plt.ylabel('difference')
plt.legend()
```

## Exercise 2

```python
def g(x):
    return np.exp(-x)*np.cos(x)+x*np.exp(-(x**2))*np.cosh(x)

def actualdv(x):
  return - np.exp(-x)*(np.cos(x)+np.sin(x)) \
         + np.exp(-(x**2))*((1-2*x**2)*np.cosh(x) + x*np.sinh(x))

x=np.linspace(0,5,20)
print(x)
```

```python
h=0.01
dfdxf=(g(x+h)-g(x))/h
print(dfdxf)
dfdxb=(g(x)-g(x-h))/h
print(dfdxb)
print(actualdv(x))
```

```python
plt.plot(x,dfdxf,'o' ,label="forward derivative")
plt.plot(x,dfdxb,'.',label="backword derivative")
plt.plot(x,actualdv(x),'-',label="Actual derivative")
plt.xlabel('x')
plt.ylabel('dfdx')
plt.legend()
```

## Exercise 3

```python
def f(x,y):
    return x**2+y**2+x*y

h = 0.01
x,y=0.5,0.2
dfdxf = (f(x+h,y)-f(x,y))/h
print(dfdxf)
```

```python
#for dfdy
def actualdv(x,y):
    return 2*y+x

x=np.linspace(0,2,20)
y=0.5
dfdyf = (f(x,y+h)-f(x,y))/h
print(dfdyf,actualdv(x,y))
```

```python
plt.plot(x,dfdyf,'-',label="forward partial differentiation")
plt.plot(x,actualdv(x,y),'o', label="differentiation using formula")
plt.xlabel('x')
plt.ylabel('pdfdxb')
plt.legend()
```
