```python
import math
from scipy import optimize
```

```python
def f(x):
    return math.cos(x) - x

def df(x):
    return -math.sin(x) - 1
```

### bisection method

```python
def bisection_method(func, a, b, tol):
    iterations = 0

    while (b - a) / 2 > tol:
        c = (a + b) / 2
        iterations += 1

        if func(c) == 0:
            return c, iterations

        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c

    root = (a + b) / 2
    return root, iterations

a = 0
b = 1
x0 = 0.5
tolerance = 1e-16

b_root, b_iter = bisection_method(f, a, b, tolerance)
```

```python
print("Custom Bisection Method Root:", b_root)
print("Bisection Iterations:", b_iter)

scipy_bisect = optimize.bisect(f, a, b)
print("SciPy Bisection Root:", scipy_bisect)
```

### newton_raphson

```python
def newton_raphson(func, dfunc, x0, tol):
    iterations = 0
    x = x0

    while True:
        x_new = x - func(x) / dfunc(x)
        iterations += 1

        if abs(x_new - x) < tol:
            break

        x = x_new

    return x_new, iterations

n_root, n_iter = newton_raphson(f, df, x0, tolerance)
```

```python
print("Custom Newton-Raphson Root:", n_root)
print("Newton-Raphson Iterations:", n_iter)

scipy_newton = optimize.newton(f, x0)
print("SciPy Newton Root:", scipy_newton)
```
