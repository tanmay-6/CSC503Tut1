```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Exercise 1

```python
demo_dict ={}
safelimit = 20
for _ in range(20):
    x = 0.1 + 5*np.random.rand()
    sum = 0
    for n in range(safelimit):
        sum += (((-1)**n)*(x**n))/(math.factorial(n))
    demo_dict[x] = sum

print(demo_dict)
```

```python
dict = {k: v for k, v in sorted(demo_dict.items(), key=lambda item: item[1])}
X = []
Y = []
for key, value in dict.items():
    X.append(key)
    Y.append(value)

plt.plot(X,Y)
```

## Exercise 2

```python
array = []
for _ in range(100):
    array.append(round(np.random.rand()*100))

print(array)
```

```python
smallest = 101
largest = -1
for i in array:
    if i <= smallest:
        smallest = i
    if i >= largest:
        largest = i

print(smallest, largest)
```
